```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

### Line plot

```python
sns.lineplot(x=df.cylinders, y=df.mpg)
```

```python
sns.lineplot(x=df.cylinders, y=df.mpg,err_style=None)
```

### Scatter with various parameter

```python
df.head(2)
```

```python
sns.scatterplot(df.acceleration,df.displacement,hue=df.origin,size=df.cylinders)
#hue is to add 3r ddimension
```

```python
plt.figure(figsize=(16,4))
sns.scatterplot(df.acceleration,df.displacement,hue=df.origin,size=df.cylinders,palette='gist_rainbow_r')
plt.show()
```

```python
m = ['*','D','d','<','>','^','+','h','H']
s=df.origin.value_counts().size
```

```python
plt.figure(figsize=(16,4))
sns.scatterplot(df.acceleration,df.displacement,hue=df.origin,markers=m[:s],style=df.origin,size=df.cylinders)
plt.show()
```

## Swarm Plot

```python
sns.violinplot(df.acceleration,orient='v',color='g')
sns.swarmplot(df.acceleration,orient='v',color='w')
plt.show()
```

```python
sns.violinplot(x=df.origin,y=df.acceleration,orient='v',color='g',inner='quartile')
sns.swarmplot(x=df.origin,y=df.acceleration,orient='v',color='r',hue=df.cylinders,alpha=.2)
plt.show()
```

## Heatmap plot

```python
dia = sns.load_dataset('diamonds')
```

```python
dia.shape
```

```python
dia.corr()
```

```python
cor=dia.corr()
```

```python
sns.heatmap(cor,annot=True,fmt='.2f',cbar=False)
```
